```python
import wandb
api = wandb.Api()

# Project is specified by <entity/project-name>
runs = api.runs("aceleo/rea_cnn_wrongkey01")
summary_list = [] 
config_list = [] 
name_list = []
a = []
for run in runs: 
    a.append(run.history()['vali_loss'].to_numpy())
    
```

```python
import numpy as np
a_np = np.array(a)
import matplotlib.pyplot as plt
plt.plot(a_np[:-1].T,c='0.85' )
plt.plot(a_np[1].T,c='0.85', label = 'wrong key' )
plt.plot(a_np[-1].T, c= 'r', label = 'right key')
plt.xlabel('Epoch')
plt.ylabel('Vali Loss')
plt.legend(loc ="upper right")
```
